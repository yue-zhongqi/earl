```python
# Settings
model = "3b"
iteration = "100"
data = "2"


from transformers import AutoTokenizer
import numpy as np
import matplotlib.pyplot as plt

tokenizer = AutoTokenizer.from_pretrained(f"Qwen/Qwen2.5-{model.upper()}-Instruct", trust_remote_code=True)
data = np.load(f"../critic_log/exp4-qwen2.5-{model}-data{data}/{iteration}.npz")
idx = 0
```

```python
responses = tokenizer(data['responses'][idx])['input_ids']
values = data['values'][idx]

n_tokens = len(responses)

decoded_responses = [tokenizer.decode([id]).replace("\n", "") for id in responses]

print(data['prompts'][idx])
print(data['responses'][idx])


# Plot the values as a line chart
plt.figure(figsize=(18, 6))
plt.plot(values[:n_tokens], marker='o', linestyle='-', color='b', label='Values')


# Add labels and title
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Line Chart of Values')
plt.legend()
plt.xticks(ticks=range(n_tokens), labels=decoded_responses, rotation=90, fontsize=5)

# Show the plot
plt.grid(True)
plt.tight_layout()  # Adjust layout to prevent label overlap
plt.show()

# for i, id in enumerate(responses):
#     # print so that values are left-aligned
#     print(f"{tokenizer.decode(id).strip():<20}: {values[i]}")
```
